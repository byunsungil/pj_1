```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re
import os

# 1. 다운로드 폴더 및 Chrome 옵션 설정 (본인 환경에 맞게 수정)
download_dir = r"C:\Users\erety\sk_13_5_1st_sungil\1st_pj_g5\새 폴더\data"
os.makedirs(download_dir, exist_ok=True)

options = Options()
prefs = {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
}
options.add_experimental_option("prefs", prefs)

# 2. ChromeDriver 경로 설정 (본인 환경에 맞게 수정)
driver_path = r"C:\chromedriver-win64\chromedriver-win64\chromedriver.exe"
service = Service(driver_path)
driver = webdriver.Chrome(service=service, options=options)

try:
    # 3. 대상 페이지 접속
    url = "https://stat.molit.go.kr/portal/cate/statMetaView.do?hRsId=58"
    driver.get(url)
    wait = WebDriverWait(driver, 30)
    time.sleep(5)  # 페이지 및 동적 콘텐츠 로딩 대기

    # 4. 다운로드 관련 <a> 태그 모두 수집
    #    다운로드 링크는 onclick 속성이 downFile 함수로 시작하는 태그입니다.
    anchors = driver.find_elements(By.CSS_SELECTOR, "a[onclick^='javascript:downFile']")
    
    for a in anchors:
        try:
            em_element = a.find_element(By.TAG_NAME, "em")
            file_text = em_element.text.strip()  # 예: "- 2023년 01월 자동차 등록자료 통계.xlsx"
            # 파일명에서 연도와 월 추출
            m = re.search(r'(\d{4})년\s*(\d{2})월', file_text)
            if m:
                year = int(m.group(1))
                month = int(m.group(2))
                # 조건: 2023년 1월부터 2025년 3월까지만
                if (year < 2023) or (year > 2025) or (year == 2025 and month > 3):
                    continue
                
                print(f"▶ 다운로드 시도: {file_text}")
                # 다운로드 링크가 화면에 보이도록 스크롤 후 클릭
                driver.execute_script("arguments[0].scrollIntoView(true);", a)
                time.sleep(1)
                driver.execute_script("arguments[0].click();", a)
                # 각 파일 다운로드 후 충분한 대기 (파일 크기 및 네트워크에 따라 조정)
                time.sleep(5)
        except Exception as e_sub:
            print("다운로드 요소 처리 오류:", e_sub)
    
    print("▶ 모든 파일 다운로드 시도 완료.")
    # 모든 다운로드가 완료될 때까지 추가 대기 (옵션)
    time.sleep(10)

except Exception as e:
    print("전체 오류 발생:", e)
    
finally:
    driver.quit()
```

```python
import pandas as pd
import os

# 1. 엑셀 파일들이 있는 폴더 경로
folder_path = r"C:\Users\erety\sk_13_5_1st_sungil\1st_pj_g5\새 폴더\data"  # 본인 경로로 수정
target_sheet = "08.차종별 등록현황(영업용)"  # 가져올 시트 이름

# 2. 해당 폴더 내 .xlsx 파일 목록 가져오기
excel_files = [file for file in os.listdir(folder_path) if file.endswith(".xlsx")]

# 3. 불러온 데이터 저장 리스트
df_list = []

for file in excel_files:
    file_path = os.path.join(folder_path, file)
    try:
        # skiprows=2로 2행 생략하고 시트 불러오기
        df = pd.read_excel(file_path, sheet_name=target_sheet, skiprows=2)

        # 상위 4행 제거 (index 기준)
        df = df.drop(df.columns[0:4], axis=1)

        # 원본 파일명 열 추가 (선택)
        df["파일명"] = file

        df_list.append(df)
        print(f"✅ 불러옴: {file}")

    except Exception as e:
        print(f"❌ 오류 발생: {file} - {e}")

combined_df = pd.concat(df_list, axis=1)
```
