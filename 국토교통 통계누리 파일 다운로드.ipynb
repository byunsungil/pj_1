```python
# 국토교통통계누리 자동 파일 다운로드 selenium


from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
import time


# ChromeOptions 설정: 다운로드 폴더 지정 (원하는 경로로 수정)
options = webdriver.ChromeOptions()
prefs = {"download.default_directory": r"C:\Users\erety\sk_13_5_1st_sungil\1st_pj_g5\새 폴더"} 
options.add_experimental_option("prefs", prefs)

# ChromeDriver 경로 설정 (본인 환경에 맞게 올바른 경로로 수정)
service = Service(r"C:\chromedriver-win64\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service, options=options)

try:
    # 1. 대상 사이트 접속
    driver.get("https://stat.molit.go.kr/portal/cate/statView.do?hRsId=58&hFormId=5498")
    
    # 2. 최대 30초까지 대기
    wait = WebDriverWait(driver, 30)
    
    # 3. 시작 기간과 마지막 기간 드롭다운 요소 로드 후 선택
    start_select_element = wait.until(EC.presence_of_element_located((By.ID, "sStart")))
    end_select_element = wait.until(EC.presence_of_element_located((By.ID, "sEnd")))
    start_select = Select(start_select_element)
    end_select = Select(end_select_element)
    start_select.select_by_value("202301")
    end_select.select_by_value("202503")
    
    # 4. 조회 버튼 클릭: onclick 속성이 "javascript:goformView();"인 버튼
    search_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@onclick='javascript:goformView();']")))
    driver.execute_script("arguments[0].scrollIntoView();", search_button)
    search_button.click()
    
    # 5. 조회 결과 로딩 대기 (필요 시 시간 조정)
    time.sleep(3)
    
    # 6. 파일 다운로드 버튼 클릭 전에 preloader(로딩 화면)가 사라질 때까지 대기
    try:
        wait.until(EC.invisibility_of_element_located((By.ID, "preLoading2")))
    except Exception:
        pass  # preloader가 없으면 넘어감
    
    file_download_btn = wait.until(EC.element_to_be_clickable((By.ID, "fileDownBtn")))
    driver.execute_script("arguments[0].scrollIntoView();", file_download_btn)
    file_download_btn.click()
    
    # 7. 파일 다운로드 팝업(또는 영역)이 열릴 때까지 잠시 대기
    time.sleep(2)
    
    # 8. Excel 아이콘 클릭: alt 속성이 "EXCEL file"인 img 태그 선택
    excel_icon = wait.until(EC.element_to_be_clickable((By.XPATH, "//img[@alt='CSV file']")))
    driver.execute_script("arguments[0].scrollIntoView();", excel_icon)
    excel_icon.click()
    
    # 9. 최종 다운로드 버튼 클릭: onclick 속성이 "download()"인 버튼
    final_download_btn = wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//button[@class='mu-btn mu-btn-primary' and @onclick='download()']")))
    driver.execute_script("arguments[0].scrollIntoView();", final_download_btn)
    driver.execute_script("arguments[0].click();", final_download_btn)
    
    # 10. 파일 다운로드 진행 시간 대기
    time.sleep(10)
    
except Exception as e:
    print("오류 발생:", e)
    
finally:
    driver.quit()
```
